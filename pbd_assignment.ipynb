   # Find the total number of patients of each disease (diagnosis) with respect to the disease state(malignant or benign) and plot in a pie chart.
   # Find and plot via pia chart the total number of patients of each disease (diagnosis) with respect to the gender(Female, Male)


```python


import pandas as pd
import numpy as np
df = pd.read_csv("jsrt_metadata.csv")
print("Data Before cleaning")
print("*********************")
print(df.isnull().sum())
print("*********************")

```

```python
df['subtlety'].replace(np.nan, df['subtlety'].astype(float).mean(), inplace=True)

df['size'].replace(np.nan, df['size'].astype(float).mean(), inplace=True)

df['x'].replace(np.nan, df['x'].astype(float).mean(), inplace=True)

df['y'].replace(np.nan, df['y'].astype(float).mean(), inplace=True)


df['position'].replace(np.nan, df['position'].value_counts().idxmax(), inplace=True)

df['diagnosis'].replace(np.nan, df['diagnosis'].value_counts().idxmax(), inplace=True)


print("Data after cleaning")
print("*********************")
print(df.isnull().sum())
print("*********************")
```

# Find the total number of patients of each disease (diagnosis) and plot in histogram.


```python

import matplotlib as plt
from matplotlib import pyplot
plt.rcParams['axes.facecolor'] = 'grey'
plt.rcParams["figure.figsize"] = (20, 10)
plt.style.use('ggplot')
df['diagnosis'].value_counts() 

plt.pyplot.xlabel("Diagnosis")
plt.pyplot.ylabel("Number of Patients")
plt.pyplot.hist(df['diagnosis'],color = 'pink',edgecolor='white')
```

# Find the total number of patients of each disease (diagnosis) with respect to the patient gender (Male, Female) and plot in a pie chart.

```python
for i in list(df['diagnosis'].unique()):
   
    print(df.loc[df['diagnosis'] == i]['state'].value_counts())
```

# Pie chart

```python
import matplotlib as plt
from matplotlib import pyplot
plt.rcParams["figure.figsize"] = (20, 5)



for i in list(df['diagnosis'].unique()):
    plt.pyplot.figure()   
    plt.pyplot.pie(df.loc[df['diagnosis'] == i]['state'].value_counts(),
                   labels=list(df.loc[df['diagnosis'] == i]['state'].value_counts().keys())
                   ,colors = ['#ff9999','#66b3ff'],wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' },
                   shadow=True,startangle=90,autopct="%1.1f%%")
    plt.pyplot.legend(title= i)
```
